### Bước 1: Nhập và lọc dữ liệu

#### 1.1: Nhập dữ liệu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Danh_sach_diem.csv')
df.columns = df.columns.str.strip()
unique_value1 = set(df['Nhom_DT1'])
unique_value2 = set(df['Nhom_DT2'])
unique_value3 = set(df['Nhom_DT3'])


print(unique_value1)
print(unique_value2)
print(unique_value3)

df

```

#### 1.2: Lọc dữ liệu

```python
df.loc[df['Nhom_DT2'] == 'Unnamed: 20', 'Nhom_DT2'] = np.nan
df.loc[df['Nhom_DT3'] == 'Unnamed: 21', 'Nhom_DT3'] = np.nan
```

```python
df.rename(columns = {'Nhom_DT2' : 'Nhom_DT1'}, inplace = True)
df.rename(columns = {'Nhom_DT3' : 'Nhom_DT2'}, inplace = True)
df
```

```python
print(df['Doi_tuong_tuyen_thang'].unique())
print(df.columns)
```

#### 1.3: Đặt lại kiểu dữ liệu cho các cột

```python
# Convert data type to numeric to handle unwanted values
df['Diem_Ngu_van'] = pd.to_numeric(df['Diem_Ngu_van'], errors='coerce').fillna(-1).astype(float)
df['Diem_Tieng_Anh'] = pd.to_numeric(df['Diem_Tieng_Anh'], errors='coerce').fillna(-1).astype(float)
df['Diem_Toan'] = pd.to_numeric(df['Diem_Toan'], errors='coerce').fillna(-1).astype(float)
df['Diem_Uu_tien'] = pd.to_numeric(df['Diem_Uu_tien'], errors='coerce').fillna(0).astype(float)
df['Tuyent_thang'] = pd.to_numeric(df['Tuyent_thang'], errors='coerce').fillna(0).astype(int)

df['Nam_sinh'] = pd.to_numeric(df['Nam_sinh'], errors='coerce').fillna(-1).astype(int)
df['Thang_sinh'] = pd.to_numeric(df['Thang_sinh'], errors='coerce').fillna(-1).astype(int)
df['Ngay_sinh'] = pd.to_numeric(df['Ngay_sinh'], errors='coerce').fillna(-1).astype(int)


df['NV1'] = pd.to_numeric(df['NV1'], errors='coerce').fillna(-1).astype(float)
df['NV2'] = pd.to_numeric(df['NV2'], errors='coerce').fillna(-1).astype(float)

# set appropriate data type to each column
df['CMND_CCCD_Ma_DDCN'] = df['CMND_CCCD_Ma_DDCN'].fillna('Unknown').astype(str)
df['DTNT'] = df['DTNT'].fillna('Unknown').astype(str)
df['Dan_toc'] = df['Dan_toc'].fillna('Unknown').astype(str)
df['Doi_tuong_tuyen_thang'] = df['Doi_tuong_tuyen_thang'].fillna("No").astype(str)
df['Ghi_chu'] = df['Ghi_chu'].fillna('None').astype(str)
df['Gioi_tinh'] = df['Gioi_tinh'].fillna('Unknown').astype(str)
df['Ho_ten_thi_sinh'] = df['Ho_ten_thi_sinh'].fillna('Unknown').astype(str)
df['Ho_va_ten_thi_sinh_khac'] = df['Ho_va_ten_thi_sinh_khac'].fillna('None').astype(str)
df['Hoc_sinh_truong'] = df['Hoc_sinh_truong'].fillna('Unknown').astype(str)
df['Huyen/Thanh_pho'] = df['Huyen/Thanh_pho'].fillna('Unknown').astype(str)
df['Huyen_TX_TP_noi_hs_hoc_cap_THCS'] = df['Huyen_TX_TP_noi_hs_hoc_cap_THCS'].fillna('Unknown').astype(str)

df['Ma_DVDKDT'] = df['Ma_DVDKDT'].fillna(-1).astype(int)




df['Nhom_DT1'] = df['Nhom_DT1'].fillna("No").astype(str)
df['Nhom_DT2'] = df['Nhom_DT2'].fillna("No").astype(str)
df['Noi_sinh_huyen_tinh'] = df['Noi_sinh_huyen_tinh'].fillna("Unknown").astype(str)
df['Noi_thuong_tru_thon_xa_huyen_tinh'] = df['Noi_thuong_tru_thon_xa_huyen_tinh'].fillna("Unknown").astype(str)
df['So_bao_danh'] = df['So_bao_danh'].fillna("Unknown").astype(str)
df['Ten'] = df['Ten'].fillna('Unknown').astype(str)
df['Tuyent_thang'] = df['Tuyent_thang'].fillna(0).astype(int)

df
```

#### 1.4: Bỏ các cột không cần thiết

```python
df.drop(columns = ['Ho_va_ten_thi_sinh_khac', 'Ten'], inplace = True)
df.loc[df['So_bao_danh'] == '110001', 'Doi_tuong_tuyen_thang'] = "No"
df.loc[df['So_bao_danh'] == '110001', 'Ghi_chu'] = "None"
df
```

### Bước 2: Phân tích

#### 2.1: Thêm cột điểm tuyển sinh

```python
df['Diem_tuyen_sinh'] = 2 * (df['Diem_Ngu_van'] + df['Diem_Toan']) + df['Diem_Tieng_Anh'] + df['Diem_Uu_tien']
df.loc[(df['Diem_Tieng_Anh'] < 0) | (df['Diem_Ngu_van'] < 0) | (df['Diem_Toan'] < 0), 'Diem_tuyen_sinh'] = -1
df
```

#### 2.2: Tính điểm trung bình theo dân tộc

#### 2.2.1: Chuẩn hoá cột dân tộc(Sửa lỗi chính tả, đánh máy,..)

```python
# Strip whitespace
df['Dan_toc'] = df['Dan_toc'].str.strip()

# Standardize 'Kinh'
df.loc[df['Dan_toc'].isin(['kinh', 'KInh', ' Kinh', 'KINH']), 'Dan_toc'] = 'Kinh'

# Standardize 'Tày'
df.loc[df['Dan_toc'].isin(['Tày', '   Tày', 'Tày', 'Tày ']), 'Dan_toc'] = 'Tày'

# Standardize 'Cơ Ho'
df.loc[df['Dan_toc'].isin(['Hồ', 'Hó']), 'Dan_toc'] = 'Cơ Ho'

# Standardize other ethnic groups
df.loc[df['Dan_toc'].isin(['Hoa', 'Hoa ']), 'Dan_toc'] = 'Hoa'
df.loc[df['Dan_toc'].isin([ "H' Mông", 'Hmong', "H'Mông", 'Hmong ']), 'Dan_toc'] = "H'mông"
df.loc[df['Dan_toc'].isin(['Mông', 'MÔNG', 'mông', 'HMông', "H'mông", 'Hmông']), 'Dan_toc'] = "H'mông"
df.loc[df['Dan_toc'].isin(['Nùng', 'NÙNG']), 'Dan_toc'] = 'Nùng'
df.loc[df['Dan_toc'].isin(['Thái', 'THÁI']), 'Dan_toc'] = 'Thái'
df.loc[df['Dan_toc'].isin(['Mường', 'Muong', 'Mường ']), 'Dan_toc'] = 'Mường'
df.loc[df['Dan_toc'] == 'Không xác định', 'Dan_toc'] = 'Unknown'
df.loc[df['Dan_toc'] == 'TÀY', 'Dan_toc'] = 'Tày'
df.loc[df['Dan_toc'] == 'DAO', 'Dan_toc'] = 'Dao'

df_dan_toc = df[df['Dan_toc'] != "Unknown"]
print(df_dan_toc['Dan_toc'].unique())
df
```

#### 2.2.2: Thực hiện các tính toán chung

```python
df_dan_toc = df_dan_toc[df_dan_toc['Diem_tuyen_sinh'] != -1]
df_grouped = df_dan_toc.groupby('Dan_toc').agg(
    So_thi_sinh = ('CMND_CCCD_Ma_DDCN', 'count'),
    So_thi_sinh_nam = ('Gioi_tinh', lambda x : (x == 'Nam').sum()),
    So_thi_sinh_nu = ('Gioi_tinh', lambda x : (x == 'Nữ').sum()),
    Diem_toan_TB = ('Diem_Toan', 'mean'),
    Diem_van_TB = ('Diem_Ngu_van', 'mean'),
    Diem_Tieng_Anh_TB = ('Diem_Tieng_Anh', 'mean'),
    Diem_tuyen_sinh_TB = ('Diem_tuyen_sinh', 'mean'),
    So_thi_sinh_duoi_20_diem = ('Diem_tuyen_sinh', lambda x : (x < 20).sum()),
    So_thi_sinh_tren_30_diem = ('Diem_tuyen_sinh', lambda x : (x > 30).sum()),
    So_thi_sinh_tren_40_diem = ('Diem_tuyen_sinh', lambda x : (x > 40).sum()),
    So_thi_sinh_tuyen_thang = ('Tuyent_thang', 'sum')
).reset_index()


df_grouped = df_grouped.round(2)
df_grouped.sort_values(by = 'So_thi_sinh', ascending = False, inplace = True)
df_grouped
```

```python
plt.figure(figsize = (16, 10))
bars = plt.bar(df_grouped['Dan_toc'], df_grouped['Diem_tuyen_sinh_TB'])
plt.title('Điểm tuyển sinh trung bình của các dân tộc')
plt.xlabel('Dân tộc')
plt.ylabel('Điểm')

plt.xticks(rotation = 45)
plt.yticks(np.arange(0, 36, 1))

for bar in bars:
    yval = bar.get_height()
    xval = bar.get_x() + bar.get_width() / 2
    plt.text(xval, yval + 0.5, yval, ha = 'center', va = 'bottom')

plt.show()
```

```python
plt.figure(figsize = (16, 5))
bars = plt.bar(df_grouped['Dan_toc'], df_grouped['Diem_Tieng_Anh_TB'])
plt.title('Điểm tiếng Anh trung bình của các dân tộc')
plt.xlabel('Dân tộc')
plt.ylabel('Điểm')

plt.xticks(rotation = 45)
plt.yticks(np.arange(0, 10.5, 0.5))

for bar in bars:
    yval = bar.get_height()
    xval = bar.get_x() + bar.get_width() / 2
    plt.text(xval, yval + 0.2, yval, ha = 'center', va = 'bottom')

plt.show()
```

```python
plt.figure(figsize = (16, 7))
bars = plt.bar(df_grouped['Dan_toc'], df_grouped['So_thi_sinh_tren_40_diem'])
plt.title('Số học sinh có điểm tuyển sinh trên 40 điểm')
plt.xlabel('Dân tộc')
plt.ylabel('Điểm')

plt.xticks(rotation = 45)
plt.yticks(np.arange(0, 201, 10))

for bar in bars:
    yval = bar.get_height()
    xval = bar.get_x() + bar.get_width() / 2
    plt.text(xval, yval + 3, yval, ha = 'center', va = 'bottom')

plt.show()
```

#### 2.2.3: Phân tích điểm người Kinh

```python
expected_value_Van = df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Ngu_van'].mean().round(2)
expected_value_Toan = df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Toan'].mean().round(2)
expected_value_Anh = df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Tieng_Anh'].mean().round(2)
expected_value_Diem_tuyen_sinh = df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_tuyen_sinh'].mean().round(2)

variance_Van = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Ngu_van'].var(), 2)
variance_Toan = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Toan'].var(), 2)
variance_Anh = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Tieng_Anh'].var(), 2)
variance_Diem_tuyen_sinh = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_tuyen_sinh'].var(), 2)

standard_deviation_Van = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Ngu_van'].std(), 2)
standard_deviation_Toan = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Toan'].std(), 2)
standard_deviation_Anh = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_Tieng_Anh'].std(), 2)
standard_deviation_Diem_tuyen_sinh = round(df_dan_toc[df_dan_toc['Dan_toc'] == 'Kinh']['Diem_tuyen_sinh'].std(), 2)

print(f"Expected Value - Văn: {expected_value_Van}")
print(f"Expected Value - Toán: {expected_value_Toan}")
print(f"Expected Value - Tiếng Anh: {expected_value_Anh}")
print(f"Expected Value - Điểm tuyển sinh: {expected_value_Diem_tuyen_sinh}")

print(f"Variance - Văn: {variance_Van}")
print(f"Variance - Toán: {variance_Toan}")
print(f"Variance - Tiếng Anh: {variance_Anh}")
print(f"Variance - Điểm tuyển sinh: {variance_Diem_tuyen_sinh}")

print(f"Standard Deviation - Văn: {standard_deviation_Van}")
print(f"Standard Deviation - Toán: {standard_deviation_Toan}")
print(f"Standard Deviation - Tiếng Anh: {standard_deviation_Anh}")
print(f"Standard Deviation - Điểm tuyển sinh: {standard_deviation_Diem_tuyen_sinh}")

df_dan_toc.head(5)
```

#### Dựa trên các thông số về kỳ vọng, phương sai và độ lệch chuẩn của các môn học và điểm tuyển sinh, có thể rút ra nhận xét chung rằng: Điểm Văn có mức trung bình cao nhất trong các môn, cho thấy đây là môn mà học sinh đạt kết quả tốt hơn so với Toán và Tiếng Anh. Điểm Toán và Tiếng Anh có trung bình thấp hơn, phản ánh sự dốt nát của học sinh ở hai môn này.Mức độ phân tán của điểm Văn là thấp nhất, đa số học sinh có điểm gần với trung bình, chủ yếu do đề Yên Bái dễ, trong khi Toán và Tiếng Anh có độ phân tán lớn hơn, cho thấy sự chênh lệch đáng kể về kết quả của các học sinh. Điểm tuyển sinh tổng thể có sự phân hóa lớn nhất, với độ lệch chuẩn cao, phản ánh sự khác biệt rõ rệt trong năng lực tổng thể của các thí sinh. Nhìn chung ở đâu cũng thế.







#### 2.3: Tính điểm trung bình theo khu vực

#### 2.3.1: Chuẩn hoá tên các huyện, thành phố

```python
df['Huyen/Thanh_pho'] = df['Huyen/Thanh_pho'].str.strip().replace({
    'Tp Yên Bái': 'TP Yên Bái',
    'Văn Yên ': 'Văn Yên',
    'Khác' : 'Unknown'
})
print(df['Huyen/Thanh_pho'].unique())
```

#### 2.3.2: Tính điểm trung bình của từng khu vực

```python
df_huyen_thanh_pho = df[df['Huyen/Thanh_pho'] != 'Unknown']

df_grouped = df_huyen_thanh_pho.groupby('Huyen/Thanh_pho').agg(
    So_thi_sinh = ('CMND_CCCD_Ma_DDCN', 'count'),
    Diem_Toan_trung_binh = ('Diem_Toan', 'mean'),
    Diem_Van_trung_binh =  ('Diem_Ngu_van', 'mean'),
    Diem_Tieng_Anh_trung_binh = ('Diem_Tieng_Anh', 'mean'),
    Diem_tuyen_sinh_trung_binh = ('Diem_tuyen_sinh', 'mean'),
    So_hoc_sinh_tuyen_thang = ('Tuyent_thang', 'sum'), 
    So_thi_sinh_duoi_20_diem = ('Diem_tuyen_sinh', lambda x : (x < 20).sum()),
    So_thi_sinh_tren_30_diem = ('Diem_tuyen_sinh', lambda x : (x > 30).sum()),
    So_thi_sinh_tren_40_diem = ('Diem_tuyen_sinh', lambda x : (x > 40).sum())
).reset_index()
df_grouped = df_grouped.round(2)
df_grouped.sort_values(by = 'Diem_tuyen_sinh_trung_binh', ascending = False)
```

```python
plt.figure(figsize=(16, 10))
bars = plt.bar(df_grouped['Huyen/Thanh_pho'], df_grouped['Diem_tuyen_sinh_trung_binh'], width = 0.4)  # Make sure to use the exact column name
plt.title('Điểm theo khu vực')
plt.xlabel('Huyện/TP')
plt.ylabel('Điểm Tuyển Sinh Trung Bình')
plt.xticks(rotation=45)

for bar in bars:
    yval = bar.get_height()
    xval = bar.get_x() + bar.get_width() / 2
    plt.text(xval, yval + 0.5, yval, ha = 'center', va = 'bottom')
plt.show()
```

#### 2.3.3: Tuyên dương các em học giỏi

```python
df_grouped = df_huyen_thanh_pho.groupby('Huyen/Thanh_pho').agg(
    Max_Toan = ('Diem_Toan', 'max'),
    HS_Toan = ('Ho_ten_thi_sinh', lambda x : x[df_huyen_thanh_pho.loc[x.index, 'Diem_Toan'].idxmax()]),
    Max_Van = ('Diem_Ngu_van', 'max'),
    HS_Van = ('Ho_ten_thi_sinh', lambda x : x[df_huyen_thanh_pho.loc[x.index, 'Diem_Ngu_van'].idxmax()]),
    Max_Anh = ('Diem_Tieng_Anh', 'max'),
    HS_Anh = ('Ho_ten_thi_sinh', lambda x : x[df_huyen_thanh_pho.loc[x.index, 'Diem_Tieng_Anh'].idxmax()]),
    Max_TS = ('Diem_tuyen_sinh', 'max'),
    HS_TS = ('Ho_ten_thi_sinh', lambda x : x[df_huyen_thanh_pho.loc[x.index, 'Diem_tuyen_sinh'].idxmax()])
).reset_index()

df_grouped.sort_values(by='Max_TS', ascending=False)
```

#### 2.3.4: Phân tích điểm huyện Trạm Tấu

```python
df_Tram_Tau = df_huyen_thanh_pho[df_huyen_thanh_pho['Huyen/Thanh_pho'] == 'Trạm Tấu']
df_grouped = df_Tram_Tau.groupby(['Dan_toc']).agg(
    So_luong = ('CMND_CCCD_Ma_DDCN', 'count'),
    So_hoc_sinh_tuyen_thang = ('Tuyent_thang', 'sum'),
    Diem_tuyen_sinh_trung_binh = ('Diem_tuyen_sinh', lambda x : x[df_Tram_Tau['Diem_tuyen_sinh'] != -1].mean())
).reset_index()

df_grouped = df_grouped.round(2)
df_grouped
```

```python
df_grouped = df_Tram_Tau.groupby('Diem_tuyen_sinh').agg(
    So_hoc_sinh = ('Diem_tuyen_sinh', lambda x : (x != -1).sum()),
).reset_index()

df_grouped.sort_values(by = 'So_hoc_sinh', ascending = False, inplace = True)
df_grouped.head(10)
```

#### 2.3.4.1: Xếp hạng học sinh giỏi Trạm Tấu(Theo điểm tuyển sinh)

```python
df_grouped = df_Tram_Tau.filter(items = ['Ho_ten_thi_sinh', 'Gioi_tinh', 'Dan_toc', 'Diem_Uu_tien', 'Diem_Ngu_van', 'Diem_Toan', 'Diem_Tieng_Anh', 'Diem_tuyen_sinh'])

df_grouped.sort_values(by = 'Diem_tuyen_sinh', ascending = False).head(20)

```

#### 2.3.4.2: Điểm trung bình 3 môn toán văn Anh.

```python
average_Van = df_Tram_Tau['Diem_Ngu_van'].mean()
average_Toan = df_Tram_Tau['Diem_Toan'].mean()
average_Tieng_Anh = df_Tram_Tau['Diem_Tieng_Anh'].mean()

print(f"Điểm văn trung bình: {average_Toan:.2f}")
print(f"Điểm toán trung bình: {average_Toan:.2f}")
print(f"Điểm tiếng Anh trung bình: {average_Tieng_Anh:.2f}")


labels = ['Điểm văn trung bình', 'Điểm toán trung bình', 'Điểm tiếng Anh']
sizes = [average_Van, average_Toan, average_Tieng_Anh]
plt.pie(sizes, labels = labels, autopct='%1.1f%%')
plt.axis('equal')

plt.show()
```

#### Nhận xét: Các em học quá dốt
